```python
import requests
from bs4 import BeautifulSoup as bs
```

# requests & BeautifulSoup

http 페이지에 요청을 보내고, 그 요청의 결과를 받아올 수 있는 함수를 가진 모듈

- 기본적으로 HTML, CSS의 구조를 아는 것이 웹 크롤링을 제대로 이해하기 위해서는 선행되어야 한다.

```python
base_url = 'https://lostark.inven.co.kr/'
```

```python
response = requests.get(base_url)

print(response.status_code) # 응답코드
print(response.text[:100]) # HTML 코드
```

```python
soup = bs(response.content, 'html.parser')

print(type(soup))
```

```python
# parsing된 BeautifulSoup 형태의 파일을 텍스트로 변환 (보기는 같아 보이지만 데이터 타입이 다르다)
# soup.prettify() == bs(response.content, 'html.parser').prettify() == response.text
print(type(soup.prettify()))
print(soup.prettify())
```

```python
# tag name, id, class 를 통해 원하는 정보를 추출한다.
```

### find & find_all

- html 태그와 css 값을 입력하여 해당 값에 해당하는 데이터를 읽어오는 기능
- find : 위에서 부터 탐색하며, 첫 번째로 발견된 값을 리턴
- find_all : 문서 전체를 탐색하여 모든 값을 리턴

```python
soup.find_all('div')
```

### select & select_one

- css 요소들인 상속, class명 id명 등을 사용할 수 있어 효율적인 선택이 가능

```python
select = soup.select_one('div > p')
```

```python
print(select.text)
```
